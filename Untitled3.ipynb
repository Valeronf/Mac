```python
kpss_test = kpss(data_f_train['Open'], regression='c')
print("KPSS Test Results")
print("KPSS-Statistic:", kpss_test[0])
print("P-Value:", kpss_test[1])
print("Lags Used:", kpss_test[2])
print("Critical Values:", kpss_test[3])

if kpss_test[1] < 0.05:
    print("P-Value is smaller than 0.05, we reject the null hypothesis and the series is non-stationary")
else:
    print("P-Value is not smaller than 0.05, we fail to reject the null hypothesis and the series is stationary")
```

```python
def acf_pacf(data=data_f_train['Open'], lag1=12, lag2=12):
    fig, ax = plt.subplots(2, 1, figsize=(15, 14), facecolor=(0.9, 0.9, 0.9))
    plot_acf(data, lags=lag1, ax=ax[0])
    plot_pacf(data, lags=lag2, ax=ax[1])
    plt.xlabel("Lags")
    ax[0].set_ylabel('ACF value')
    ax[0].grid(axis='x')
    ax[1].set_ylabel('PACF value')
    ax[1].grid(axis='x')
    plt.show()
```

```python
arma1_0_model = ARIMA(data_f_train['Open'], order=(1, 0, 0)).fit()
arma1_7_model = ARIMA(data_f_train['Open'], order=(1, 0, 7)).fit()
arma1_14_model = ARIMA(data_f_train['Open'], order=(1, 0, 14)).fit()
arma8_0_model = ARIMA(data_f_train['Open'], order=(8, 0, 0)).fit()
arma8_8_model = ARIMA(data_f_train['Open'], order=(8, 0, 8)).fit()
arma8_28_model = ARIMA(data_f_train['Open'], order=(8, 0, 28)).fit()
arma12_0_model = ARIMA(data_f_train['Open'], order=(12, 0, 0)).fit()
arma12_14_model = ARIMA(data_f_train['Open'], order=(12, 0, 14)).fit()
arma12_28_model = ARIMA(data_f_train['Open'], order=(12, 0, 28)).fit()
```

```python
models = [
        (arma1_0_model, 1, 0),
        (arma1_7_model, 1, 7),
        (arma1_14_model, 1, 14),
        (arma8_0_model, 8, 0),
        (arma8_8_model, 8, 8),
        (arma8_28_model, 8, 28),
        (arma12_0_model, 12, 0),
        (arma12_14_model, 12, 14),
        (arma12_28_model, 12, 28)
        ]
```

```python
def metrics_arma(model, p, q):
    rss = np.sum(model.resid**2)
    r2 = 1 - rss / np.sum((model.model.endog - np.mean(model.model.endog))**2)
    aic = model.aic
    dw = durbin_watson(model.resid)

    if (q == 0):
        print("AR({}):\n R-squared: {},  RSS: {},  Akaike: {},  Durbin-Watson: {}".format(p, r2, rss, aic, dw))
    else:
        print("ARMA({}, {}):\n R-squared: {},  RSS: {},  Akaike: {},  Durbin-Watson: {}".format(p, q, r2, rss, aic, dw))


    return {
        'Model': f'ARMA({p}, {q})' if q != 0 else f'AR({p})',
        'R-squared': r2,
        'RSS': rss,
        'AIC': aic,
        'Durbin-Watson': dw
    }
```

```python
def models_comp(models):
    results = [metrics_arma(model, p, q) for model, p, q in models]
    return pd.DataFrame(results)
```

```python
results_df = models_comp(models)
results_df
```

```python
def best_models(results_df=results_df):
    return pd.DataFrame([results_df.loc[results_df['R-squared'].idxmax()], results_df.loc[results_df['RSS'].idxmin()], results_df.loc[results_df['AIC'].idxmin()], results_df.iloc[(results_df['Durbin-Watson'] - 2).abs().idxmin()]],
                           index=['R-squared', 'RSS', 'AIC', 'Durbin-Watson'])
```

```python
best_models()
```
